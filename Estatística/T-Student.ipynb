```python
import matplotlib.pyplot as fig
import xlrd
import numpy as ny
from scipy import stats as st
dad1 = [10, 11, 10, 9, 8, 9, 10, 11, 12, 9, 10, 8]
dad2 = [5, 6, 9, 10, 8, 5, 4, 5, 6, 7, 8, 5]

fig.subplot(211)
fig.plot(dad1, '-k') # Linha
fig.plot(dad2, '--k') # Linha Tracejado
fig.title('Duas amostras (dad1, dad2)')
ax = fig.subplot(212)
ax.boxplot([dad1, dad2])
ax.set_xticklabels(['dad1', 'dad2'])
print(st.ttest_ind(dad1, dad2))
```

    Teste T-student para duas amostras

```python
excel = xlrd.open_workbook("ibovespa.xls")

petr = excel.sheet_by_name("PETR3")
vale = excel.sheet_by_name("VALE3")

linPetr = petr.nrows
colPetr = petr.ncols
dadosPetr = []

linVale = vale.nrows
colVale = vale.ncols
dadosVale = []


for i in range(colPetr):
    coluna1 = petr.col_values(i)
    dadosPetr.append(coluna1)
    coluna2 = vale.col_values(i)
    dadosVale.append(coluna2)
    
dadosPetr = ny.array(dadosPetr)
dadosVale = ny.array(dadosVale)

retorno1 = (dadosPetr[1, 1:linPetr] - dadosPetr[1, 0:linPetr-1]) / dadosPetr[1, 0:linPetr-1]
retorno2 = (dadosVale[1, 1:linVale] - dadosVale[1, 0:linVale-1]) / dadosVale[1, 0:linVale-1]

# Teste T
print(st.ttest_ind(retorno1, retorno2))
fig.subplot(311)
fig.plot(retorno1, '-k')
fig.title("Retorno PETR3")

fig.subplot(312)
fig.plot(retorno2, '-k')
fig.title("Retorno VALE3")

ax = fig.subplot(313)
ax.boxplot([retorno1, retorno2])
ax.set_xticklabels(["PETR3", "VALE3"])
```
