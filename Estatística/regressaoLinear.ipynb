```python
import matplotlib.pyplot as fig
import xlrd
import numpy as ny
from scipy import stats

excel = xlrd.open_workbook("ibovespa.xls")

ibov = excel.sheet_by_name("Ibov")
petr = excel.sheet_by_name("PETR3")
abev = excel.sheet_by_name("ABEV3")
bidi = excel.sheet_by_name("BIDI4")

linIbov = ibov.nrows
colIbov = ibov.ncols
dadosIbov = []

linAbev = abev.nrows
colAbev = abev.ncols
dadosAbev = []

linBidi = bidi.nrows
colBidi = bidi.ncols
dadosBidi = []

linPetr = petr.nrows
colPetr = petr.ncols
dadosPetr = []

# Coloca todas os valores de uma coluna em uma lista
for i in range(colIbov):
    coluna1 = ibov.col_values(i)
    dadosIbov.append(coluna1)
    
    coluna2 = abev.col_values(i)
    dadosAbev.append(coluna2)
    
    coluna3 = bidi.col_values(i)
    dadosBidi.append(coluna3)
    
    coluna4 = petr.col_values(i)
    dadosPetr.append(coluna4)
    
# Transforma lista em vetor
dadosIbov = ny.array(dadosIbov)
dadosAbev = ny.array(dadosAbev)
dadosBidi = ny.array(dadosBidi)
dadosPetr = ny.array(dadosPetr)

# Regressão Linear
beta, beta0, rValue, pValue, stdErr = stats.linregress(dadosIbov[1], dadosAbev[1])
print("Beta: {:.6f}, Beta0: {:.6f}, R_Value: {:.6f}, P_Value: {:.6f}, stdErr: {:.6f} ".format(beta, beta0, rValue, pValue, stdErr))
print("y = {:.6f}x - {:.6}".format(beta, beta0))

yLin = beta*dadosIbov[1] + beta0
fig.plot(dadosIbov[1], yLin, '-k', dadosIbov[1], dadosAbev[1], 'ok') # -k = linha, ok = circulos
fig.xlabel("IBOVESPA")
fig.ylabel("AMBEV")
```

```python
# Regressão Linear entre Petrobras e Ibovespa
beta, beta0, rValue, pValue, stdErr = stats.linregress(dadosIbov[1], dadosPetr[1])
print("Beta: {:.6f}, Beta0: {:.6f}, R_Value: {:.6f}, P_Value: {:.6f}, stdErr: {:.6f} ".format(beta, beta0, rValue, pValue, stdErr))
print("y = {:.6f}x  {:.6}".format(beta, beta0))

yLin = beta*dadosIbov[1] + beta0
fig.plot(dadosIbov[1], yLin, '-k', dadosIbov[1], dadosPetr[1], 'ok') # -k = linha, ok = circulos
fig.xlabel("IBOVESPA")
fig.ylabel("PETROBRAS")
```

```python
# Regressão Linear entre Banco Inter e Ibovespa
beta, beta0, rValue, pValue, stdErr = stats.linregress(dadosIbov[1], dadosBidi[1])
print("Beta: {:.6f}, Beta0: {:.6f}, R_Value: {:.6f}, P_Value: {:.6f}, stdErr: {:.6f} ".format(beta, beta0, rValue, pValue, stdErr))
print("y = {:.6f}x  {:.6}".format(beta, beta0))

yLin = beta*dadosIbov[1] + beta0
fig.plot(dadosIbov[1], yLin, '-k', dadosIbov[1], dadosBidi[1], 'ok') # -k = linha, ok = circulos
fig.xlabel("IBOVESPA")
fig.ylabel("BANCO INTER")
```
