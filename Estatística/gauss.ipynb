```python
import matplotlib.pyplot as fig
import random as rd
from scipy.stats import norm
import statistics as st
import numpy as ny

aleat = []
for i in range(1000):
    x = rd.gauss(0, 1)
    aleat.append(x)

fig.subplot(211)
fig.plot(aleat, color="black")
fig.xlabel("Eixo Horizontal")
fig.ylabel("Números aleatórios entre -1 e 1")

fig.subplot(212)
fig.hist(aleat, bins=20, color="blue", density=True)
fig.xlabel("Classes")
fig.ylabel("Frequência")

xmin, xmax = fig.xlim() # Get or set the x limits of the current axes.
media = st.mean(aleat)
desvio = st.pstdev(aleat)
eixoX = ny.linspace(xmin, xmax, 100)
eixoY = norm.pdf(eixoX, media, desvio)
eixoY2 = norm.cdf(eixoX, media, desvio) # Função de desindade acumulada
fig.plot(eixoX, eixoY, color="Black")
fig.plot(eixoX, eixoY2, color="Green")
```
