```python
import xlrd
import seaborn as sns
import matplotlib.pyplot as fig
import yfinance as yf
import numpy as ny

wb = xlrd.open_workbook('EstatDad.xls')
plan = wb.sheet_by_name('Plan3')

lin = plan.nrows
col = plan.ncols

dados = []

for i in range(col):
    coluna = plan.col_values(i)
    dados.append(coluna)

sns.boxplot(data=dados)
sns.stripplot(data=dados, marker='o', color="black")
```

```python
fig.figure()
fig.subplot(221)
sns.distplot(dados[0], color="black")
fig.ylabel("AMOSTRRA 1", fontsize=14)

fig.subplot(222)
sns.distplot(dados[1], color="black")
fig.ylabel("AMOSTRRA 2", fontsize=14)

fig.subplot(223)
sns.distplot(dados[2], color="black")
fig.ylabel("AMOSTRRA 3", fontsize=14)

fig.subplot(224)
sns.distplot(dados[3], color="black")
fig.ylabel("AMOSTRRA 4", fontsize=14)
```

```python
fig.figure()
fig.subplot(221)
sns.violinplot(dados[0], color='grey')
fig.ylabel("Amostra 1", fontsize=14)

fig.subplot(222)
sns.violinplot(dados[1], color='grey')
fig.ylabel("Amostra 2", fontsize=14)

fig.subplot(223)
sns.violinplot(dados[2], color='grey')
fig.ylabel("Amostra 3", fontsize=14)

fig.subplot(224)
sns.violinplot(dados[3], color='grey')
fig.ylabel("Amostra 4", fontsize=14)
```

```python
tickersList = ["PETR3.SA", "VALE3.SA","^BVSP"]

# Downloading the data for the file, only adjusted close
data = yf.download(tickersList, period="1y")["Adj Close"]

petrobras = data["PETR3.SA"]
vale = data["VALE3.SA"]
ibov = data["^BVSP"]
```

```python
retPetrobras = petrobras.pct_change()
retVale = vale.pct_change()
retIbov = ibov.pct_change()

eixoX = ny.arange(len(petrobras))
eixoRet = ny.arange(len(retIbov))



fig.figure(figsize=(12,8))
ax1 = fig.subplot(111)
ax1.plot(eixoX, ibov, '-k')
ax1.set_xlabel("Dias", fontsize=14)
ax1.set_ylabel("Ibovespa")
ax1_segundo = ax1.twinx()
ax1_segundo.plot(eixoX, petrobras, '--k', eixoX, vale, '.-k')
```

```python
fig.figure(figsize=(8,6))
fig.plot(eixoRet, retIbov, '-k')
fig.title("Retorno IBOVESPA")
```

```python
fig.figure(figsize=(8,6))
fig.plot(eixoRet, retPetrobras, '-k')
fig.title("Retorno Petrobras")
```

```python
fig.figure(figsize=(8,6))
fig.plot(eixoRet, retVale, '-k')
fig.title("Retorno VALE")
```

```python
ax3 = fig.subplot(331)
ax3 = sns.distplot(retIbov, color='black')
ax3 = fig.title("Densidade de retorno ibovespa")

ax4 = fig.subplot(332)
ax4 = sns.distplot(retPetrobras, color='black')
ax4 = fig.title("Densidade de retorno Petrobras")

ax5 = fig.subplot(333)
ax5 = sns.distplot(retVale, color='black')
ax5 = fig.title("Densidade de retorno Vale")
```
