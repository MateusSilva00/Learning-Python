```python
import matplotlib.pyplot as fig
import random as rd

aleat = []
for i in range(100):
    x = rd.randint(1, 60) # Gera números inteiros pseudoaleatórios de 1 a 60
    aleat.append(x)    
fig.xlabel("Eixo Horizontal"); fig.ylabel("Números Pseudoaleatórios de 1 a 60"); fig.plot(aleat);
```

```python
alear = []
for i in range(100):
    y = rd.random()
    alear.append(y)
    
fig.xlabel("Eixo Horizontal"); fig.ylabel("Números Pseudoaleatórios de 0 a 1"); fig.plot(alear)
```

```python
alerb = []
for j in range(100):
    z = rd.uniform(-1 , 1)
    alerb.append(z)
fig.xlabel("Eixo Horizontal"); fig.ylabel("Números Pseudoaleatórios de -1 a 1"); fig.plot(alerb)
```
