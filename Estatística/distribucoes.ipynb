    Distribuição Uniforme

```python
import matplotlib.pyplot as fig
import random as rd
aleat = []

for i in range(1000):
    x = rd.uniform(-1, 1)
    aleat.append(x)

fig.subplot(211); fig.plot(aleat, color="black"); fig.xlabel("Eixo Horizontal"); fig.ylabel("Número pseudoaleatórios entre -1 e 1");
fig.subplot(212); fig.hist(aleat, bins=20, color="gray", density=True); fig.xlabel("Classes"); fig.ylabel("Frequência")
```

    Distribuição Gaussiana

```python
alear = []
for j in range(10000):
    y = rd.gauss(0, 1) #Distribuição normal em Z, onde a média é 0 e o desvio padrão é 1
    alear.append(y)

fig.subplot(211); fig.plot(alear, color="green"); fig.xlabel("Eixo Horizontal"); fig.ylabel("Números aleatórios entre -1 e 1");
fig.subplot(212); fig.hist(alear, bins=20, color="purple", density=True); fig.xlabel("Classes"); fig.ylabel("Frequência")
```

    Distribuição Gama

```python
alearb = []
for k in range(1000):
    z = rd.gammavariate(1,2)
    alearb.append(z)

fig.subplot(211)
fig.plot(alearb, color="black")
fig.xlabel("Eixo Horizontal")
fig.ylabel("Número pseudoaleatórios entre -1 e 1")

fig.subplot(212)
fig.hist(alearb, bins=20, color="gray", density=True)
fig.xlabel("Classes")
fig.ylabel("Frequência")
```
