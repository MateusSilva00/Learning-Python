Vinte opções da Petrobras

```python
preco = []
for i in range(20):
    x = random.uniform(1,2)
    preco.append(x)
```

```python
preco = ny.array(preco)
retorno = (preco[1:20] - preco[0:19]) / preco[0:19]
```

```python
t = ny.arange(0, 19)
fig.plot(t, retorno)
fig.title("Retorno das opções de PTBR4")
fig.xlabel("Tempo")
fig.ylabel("Retorno")
```

```python
import math
import statistics as st
media = st.mean(retorno)
desvio = st.pstdev(retorno)
otimista = media + 2 * desvio/math.sqrt(19)
pessimista = media - 2 * desvio/math.sqrt(19)
print("Otimista = {:.2f}, Pessimista = {:.2f} ".format(otimista, pessimista))
```

```python
tempo = ny.arange(0, 20)
fig.subplot(211)
fig.plot(tempo, preco)

fig.subplot(212)
fig.plot(t, retorno)
```
