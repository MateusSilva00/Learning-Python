```python
import matplotlib.pyplot as fig
x = [0, 1, 2, 3]
y = [2, 5, 8, 10]
z = [3, 3, 1, 2]
w = [10, 3, 2, 1]
fig.plot(x,y, '-k', x, z,'-or', x, w, '-*b', markersize=12)
fig.xlabel("Eixo x")
fig.ylabel("Eixo y")
fig.title("Gráfico teste")
fig.grid(True)
fig.text(x=2, y=2.5, s='Curva Z', fontsize=16)
```

```python
x = [0,1,2,3,4,5,6]
y = [2,5,8,10,8,3,1]
z = [-2,-1,3,4,5,0,-1]

fig.figure()
ax = fig.subplot(111)
g1 = ax.plot(x,y, '-*k', markersize=14)
g2 = ax.plot(x,z, '-vk', markersize=14)
ax.legend(['y', 'z'], loc='upper right')
ax.set_xlim(0,6)
ax.set_xlabel("Eixo x", fontsize=14, weight="bold")
ax.set_ylabel("Eixo y", fontsize=14, weight="bold")
ax.set_ylim(-10, 15)
ax.set_xticks(x)
ax.set_yticks([-2,-1.5,-1,-0.5,0, 0.5, 1, 1.5,2,2.5,3,3.5,4,5,6,8,10])
ax.grid(True)
ax.set_title("++++++++++++++++++ GRÁFICO COM PROPRIEDADES DOS EXISO ++++++++++++++++++")
ax.text(x=5, y=4, s='var y', fontsize=12, color='k', weight='bold')
```
